### Classificação de bons e maus pagadores com Deep Learning

#### Importação das bibliotecas

```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_transformer
```

#### Importação dos dados

```python
df = pd.read_csv('Credit2.csv', sep=';')
```

```python
df.head()
```

#### Pré processamento

```python
#separação das variáveis
x = df.iloc[:,1:10].values
#classe
y = df.iloc[:,10].values
```

```python
#df transformado em array
x
```

```python
#label encoder coluna checking_status
#atribui valores de 0 a 3
labelencoder = LabelEncoder()
x[:,0] = labelencoder.fit_transform(x[:,0])
```

```python
#array após transformação
x
```

```python
#one hot encoder coluna credit_history
#adiciona 1 coluna para cada opção
onehotencoder = make_column_transformer((OneHotEncoder(categories='auto', sparse=False), [1]), remainder='passthrough')
x = onehotencoder.fit_transform(x)
```

```python
x
```

```python
### exclusão da variável para evitar a dummy variable trap
x = x[:,1:]
```

```python
#label encoder com a classe
labelencoder_Y = LabelEncoder()
y = labelencoder_Y.fit_transform(y)
```

```python
y
```

```python
#separação em treino e teste
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
#feature scalling, padronização z-score
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

#### Criação do modelo

```python
classifier = Sequential()
classifier.add(Dense(units=6, kernel_initializer = 'uniform', activation='relu', input_dim=12))
classifier.add(Dense(units=6, kernel_initializer = 'uniform', activation='relu'))
classifier.add(Dense(units=1, kernel_initializer = 'uniform', activation='sigmoid'))
classifier.compile(optimizer = 'adam', loss='binary_crossentropy', metrics=['accuracy'])
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
y_pred = classifier.predict(X_test)
#para armazenar em True ou False ao invés de probabilidade
y_pred = (y_pred>0.5)
```

```python
y_pred
```

```python
#matriz de confusão para verificação da quantidade de acertos do modelo
cm = confusion_matrix (y_test, y_pred)
cm
```
