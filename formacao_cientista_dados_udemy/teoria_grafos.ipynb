#### Teoria dos grafos

```python
from igraph import Graph
from igraph import plot
```

```python
#definição dos grafos com as arestas
grafo1=Graph(edges = [(0,1), (1,2), (2,3), (3,0)], directed=True)

#definição do rótulo de cada vértice
grafo1.vs['label'] = range(grafo1.vcount()) #contagem de cada vértice é o label dele
print(grafo1)
```

```python
#grafico 
plot(grafo1, bbox=(0,0,300,300))
```

```python
grafo2=Graph(edges=[(0,1), (1,2), (2,3),(3,0),(1,1)], directed=True)
#adicionando o vértice isolado
grafo2.add_vertex(5)
grafo2.vs['label']=range(grafo2.vcount())
plot(grafo2, bbox=(0,0,300,300))
```
